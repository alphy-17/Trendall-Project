```python
import docx
import re

def extract_artifact_types(doc_path):
    # Load the document
    doc = docx.Document(doc_path)

    # Extract text from each paragraph in the document
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)

    # Join all text into a single string with two newline characters between paragraphs for clear separation
    full_text = "\n\n".join(full_text)

    # Regular expression pattern to match artifact type names
    pattern = r"(?<=\n\n)([A-Z][a-z\s\-\(\),]*(?:\([a-z0-9\s]+\))?)(?=\n\n)"

    # Find all occurrences of the pattern
    artifact_types = re.findall(pattern, full_text)

    # Initialize a set to store unique artifact types
    unique_artifact_types = set()

    # Filter artifact types based on length criteria and add them to the set if they are unique
    for artifact in artifact_types:
        if len(artifact) < 40:
            unique_artifact_types.add(artifact)

    return unique_artifact_types

# Path to the document
doc_path = "/content/Finalised Text Extraction sample.docx"  # Replace this with the actual path to your document

# Extract artifact types
artifact_types = extract_artifact_types(doc_path)
count = 0
# Print the extracted and unique artifact types
for artifact in sorted(artifact_types):  # Sorting before printing for readability
    print(artifact)
    count+=1

print(count)
```

```python
import re
import docx
def extract_publications(doc_path):
    # Load the document
    doc = docx.Document(doc_path)

    # Extract text from each paragraph in the document
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)

    # Join all text into a single string
    full_text = " ".join(full_text)

    # Define a regex pattern to capture common publication reference patterns
    pattern = r"(\bPP\b|\bPPSupp\b|\bPAdd\b|\bLCS\b|\bCVA\b|Suppl\.?\s*Ill|ArchReps|Art Journal|Apollo|Cat\.?),\s*(no\.|p\.|pp\.|vol\.|fig\.|pi\.|ill\.)\s*[\w\s-]+(?:,\s*(?:no\.|p\.|pp\.|vol\.|fig\.|pi\.|ill\.)\s*[\w\s-]+)*(?:;\s*|$)"

    # Find all occurrences of the pattern
    publications = re.findall(pattern, full_text)

    # Process to remove duplicates and clean data
    unique_publications = set(publications)  # Convert list to set to remove duplicates
    cleaned_publications = [pub[0].strip() for pub in unique_publications]  # Clean whitespace

    return cleaned_publications
```

```python
import json
import torch
from transformers import BertTokenizerFast, BertForTokenClassification, Trainer, TrainingArguments
from sklearn.model_selection import train_test_split

# Function to load data from a JSONL file
def load_data(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        return [json.loads(line.strip()) for line in file]
```

```python
# Function to tokenize data and align labels
def tokenize_and_align_labels(data, tokenizer, label_dict):
    tokenized_inputs = tokenizer([example['text'] for example in data], truncation=True, padding=True, max_length=512, return_offsets_mapping=True)
    labels = []

    for i, example in enumerate(data):
        doc_labels = [-100] * len(tokenized_inputs['input_ids'][i])  # Use -100 for ignored tokens
        offset_mapping = tokenized_inputs['offset_mapping'][i]

        for idx, mapping in enumerate(offset_mapping):
            if mapping[0] == mapping[1]:
                continue
            token_start, token_end = mapping
            for start, end, label in example['label']:
                if token_start >= start and token_end <= end:
                    doc_labels[idx] = label_dict[label]
                    break

        labels.append(doc_labels)

    tokenized_inputs['labels'] = labels
    return tokenized_inputs
```

```python
# Define the dataset class
class NERDataset(torch.utils.data.Dataset):
    def __init__(self, encodings):
        self.encodings = encodings

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items() if key != 'offset_mapping'}
        return item

    def __len__(self):
        return len(self.encodings['input_ids'])
```

```python
# Load and split data
data = load_data('/content/admin.jsonl')
train_val_data, test_data = train_test_split(data, test_size=0.1, random_state=42)
train_data, val_data = train_test_split(train_val_data, test_size=0.1111, random_state=42)

# Set up the tokenizer and label dictionary
tokenizer = BertTokenizerFast.from_pretrained('bert-base-cased')
label_dict = {'REFERENCE': 1, 'IMAGE_DESCRIPTION': 2, 'O': 0}

# Tokenize and align labels
train_encodings = tokenize_and_align_labels(train_data, tokenizer, label_dict)
val_encodings = tokenize_and_align_labels(val_data, tokenizer, label_dict)
test_encodings = tokenize_and_align_labels(test_data, tokenizer, label_dict)

# Create datasets
train_dataset = NERDataset(train_encodings)
val_dataset = NERDataset(val_encodings)
test_dataset = NERDataset(test_encodings)
```

```python
import json
import torch
from transformers import BertTokenizerFast, BertForTokenClassification, Trainer, TrainingArguments
from sklearn.model_selection import train_test_split
import numpy as np

# Load the tokenizer
tokenizer = BertTokenizerFast.from_pretrained('bert-base-cased')

# Load the saved model from the directory path
model_path = '/content'
model = BertForTokenClassification.from_pretrained(model_path)

# Define a function to tokenize text for prediction
def tokenize_for_prediction(text, tokenizer):
    tokenized_inputs = tokenizer(text, truncation=True, padding=True, max_length=512, return_tensors="pt")
    return tokenized_inputs

# Sample text
new_text = """
3
Palermo 2198 (old no. 3480), from Gela. Rim broken off; actual ht. 32-5. PLATE 1 b,c
LCS, p. 203, no. 28a, pi. 80,1 (reverse); Suppl. Ill, p. 99, no. 51.
[a) Standing draped woman, resting r. arm on pillar, Eros beside seated woman looking
into mirror, half-draped youth; bust of silen, above, [b) draped woman with fillet in r. hand,
young satyr bending forward with phiale in r. hand.
"""

# Tokenize the new text for prediction
new_encodings = tokenize_for_prediction(new_text, tokenizer)

# Get predictions for the new text
model.eval()
with torch.no_grad():
    outputs = model(**new_encodings)

# Decode the predictions
label_ids = torch.argmax(outputs.logits, dim=2).squeeze().tolist()
tokens = tokenizer.convert_ids_to_tokens(new_encodings['input_ids'].squeeze())

# Initialize variables to store the extracted REFERENCE and IMAGE_DESCRIPTION
reference = ""
image_description = ""
current_label = None
current_chunk = ""

# Process the predictions to extract references and image descriptions
label_map = {1: 'REFERENCE', 2: 'IMAGE_DESCRIPTION'}

print("Tokens and their predicted labels:")
for token, label_id in zip(tokens, label_ids):
    label = label_map.get(label_id, 'O')  # Default to 'O' if label_id is not in label_map
    print(f"Token: {token}, Label: {label}")  # Debug statement
    if label == 'O':
        if current_label == 'REFERENCE':
            reference += current_chunk + " "
        elif current_label == 'IMAGE_DESCRIPTION':
            image_description += current_chunk + " "
        current_chunk = ""
        current_label = None
    else:
        if current_label and current_label != label:
            if current_label == 'REFERENCE':
                reference += current_chunk + " "
            elif current_label == 'IMAGE_DESCRIPTION':
                image_description += current_chunk + " "
            current_chunk = ""
        current_chunk += token.replace("##", "") + " "
        current_label = label

# Ensure the last chunk is added to the respective category
if current_chunk:
    if current_label == 'REFERENCE':
        reference += current_chunk
    elif current_label == 'IMAGE_DESCRIPTION':
        image_description += current_chunk

# Print the extracted REFERENCE and IMAGE_DESCRIPTION
print("REFERENCE:", reference.strip())
print("IMAGE_DESCRIPTION:", image_description.strip())
```
